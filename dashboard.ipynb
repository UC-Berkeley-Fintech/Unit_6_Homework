# San Francisco Rental Prices Dashboard

In this notebook, you will compile the visualizations from the previous analysis into functions that can be used for a Panel dashboard.

```python
# imports
import panel as pn
pn.extension('plotly')
import plotly.express as px
import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt
import os
from pathlib import Path
from dotenv import load_dotenv
```

### WARNING! - DO NOT FORGET TO PROVIDE YOUR OWN env FILE.

```python
# Read the Mapbox API key
load_dotenv()
map_box_api = os.getenv("mapbox")
px.set_mapbox_access_token(map_box_api)
```

# Import Data

```python
# Import the necessary CSVs to Pandas DataFrames
file_path = Path("Data/sfo_neighborhoods_census_data.csv")
sfo_data = pd.read_csv(file_path, index_col="year")

loc_path = Path('Data/neighborhoods_coordinates.csv')
locations_df = pd.read_csv(loc_path)
```

- - -

## Panel Visualizations

In this section, you will copy the code for each plot type from your analysis notebook and place it into separate functions that Panel can use to create panes for the dashboard. 

These functions will convert the plot object to a Panel pane.

Be sure to include any DataFrame transformation/manipulation code required along with the plotting code.

Return a Panel pane object from each function that can be used to build the dashboard.

Note: Remove any `.show()` lines from the code. We want to return the plots instead of showing them. The Panel dashboard will then display the plots.

```python
# Define Panel Visualization Functions
def housing_units_per_year():
    """Housing Units Per Year."""
    
    # YOUR CODE HERE!
    housing_units = sfo_data.groupby('year')['housing_units'].mean()
    fig_avg_house_per_year = plt.figure()
    
    housing_units_plot = housing_units.plot(kind='bar', ylim=(370000, 387500))
    return fig_avg_house_per_year

def average_gross_rent():
    """Average Gross Rent in San Francisco Per Year."""
    
    # YOUR CODE HERE!
    avg_price_sft_rent = sfo_data.groupby('year')['sale_price_sqr_foot', 'gross_rent'].mean()
    
    fig_avg_gross_avg_rent_sqft = plt.figure()
    avg_price_sft_rent_plot = plt.plot(avg_price_sft_rent.index, avg_price_sft_rent['gross_rent'], color='orange')
    plt.xlabel('Year')
    plt.ylabel('Average Gross Rent')
    plt.title('Average Gross Rent by Year')
   # plt.xlim(avg_price_sft_rent.index[0], avg_price_sft_rent.index[-1])
    return fig_avg_gross_avg_rent_sqft

def average_sales_price():
    """Average Sales Price Per Year."""
    
    # YOUR CODE HERE!
    avg_price_sft_rent = sfo_data.groupby('year')['sale_price_sqr_foot', 'gross_rent'].mean()
    fig_avg_gross_avg_sale_sqft = plt.figure()
    avg_sales_plot = plt.plot(avg_price_sft_rent.index, avg_price_sft_rent['sale_price_sqr_foot'], color='green')
    plt.xlabel('Year')
    plt.ylabel('Average Sale Price per Sqft')
    plt.title('Average Sale Price per Sqft by Year')
    return fig_avg_gross_avg_sale_sqft


def average_price_by_neighborhood():
    """Average Prices by Neighborhood."""
    
    # YOUR CODE HERE!
    avg_nb_values = sfo_data.groupby(['year', 'neighborhood']).mean().reset_index()
    avg_prices_plot = avg_nb_values.hvplot.line(
    x='year',
    y='sale_price_sqr_foot',
    xlabel= 'Year',
    ylabel= 'Avg Sale Price per Sq Ft',
    groupby='neighborhood',
    width=600
    ).opts(yformatter='%0f')
    avg_prices_plot = pn.pane.HoloViews(avg_prices_plot)
    return avg_prices_plot



def top_most_expensive_neighborhoods():
    """Top 10 Most Expensive Neighborhoods."""

    # YOUR CODE HERE!
    top_ten = sfo_data.groupby(['neighborhood']).mean().sort_values(by='sale_price_sqr_foot', ascending=False).head(10)
    
    top_ten_plot = top_ten.hvplot.bar(
    x='neighborhood',
    y='sale_price_sqr_foot',
    xlabel= 'Neighborhood',
    ylabel= 'Avg Sale Price per Sqft',
    rot=90,
    title='Top 10 expensive neighborhoods',
    height=400
    )
    top_ten_plot = pn.pane.HoloViews(top_ten_plot)
    return top_ten_plot

def most_expensive_neighborhoods_rent_sales():
    """Comparison of Rent and Sales Prices of Most Expensive Neighborhoods."""   
    
    # YOUR CODE HERE!
    avg_nb_values = sfo_data.groupby(['year', 'neighborhood']).mean().reset_index()
    
    comp_plot = avg_nb_values.hvplot.bar(
    x='year',
    y=['sale_price_sqr_foot', 'gross_rent'],
    xlabel= 'Neighborhood',
    ylabel= 'Avg Sale Price per Sqft',
    rot=90,
    title='Top 10 expensive neighborhoods',
    groupby='neighborhood',
    height=400
    )
    comp_plot = pn.pane.HoloViews(comp_plot)
    return comp_plot


def neighborhood_map():
    """Neighborhood Map."""

    # YOUR CODE HERE!
    nb_df = sfo_data.groupby('neighborhood').mean()
    nb_df.reset_index(inplace=True)
    combined_df = pd.concat([locations_df, nb_df], axis='columns', join='inner')
    combined_df = combined_df.drop(['neighborhood'], axis=1)
    
    map_scatter = px.scatter_mapbox(combined_df,
    lat='Lat',
    lon='Lon',
    size='sale_price_sqr_foot',
    color='gross_rent',
    color_continuous_scale=px.colors.cyclical.IceFire,
    size_max=15,
    zoom=11,
    hover_name='Neighborhood',
    title='Averange Sale Price Per Square Foot and Gross Rent in San Francisco',
    )
    return map_scatter
```

## Panel Dashboard

In this section, you will combine all of the plots into a single dashboard view using Panel. Be creative with your dashboard design!

```python
# Create a Title for the Dashboard
title = '###Real Estate Analysis of San Francisco from 2010-2016'
explainer_text = 'This dashboard represents a historical analysis of real estate conditions in the San Fransisco area.'

# Create a tab layout for the dashboard
welcome_tab = pn.Column(title , explainer_text, neighborhood_map())

# '###San Francisco housing market overview',
market_analysis_tab = pn.Row(
    housing_units_per_year(),
    average_gross_rent(),
    average_sales_price()
)

neighborhood_analysis_tab = pn.Column(
    '###San Francisco Top 10 Neighborhoods',
    average_price_by_neighborhood(), 
    top_most_expensive_neighborhoods(), 
    most_expensive_neighborhoods_rent_sales()
    )

# Create the dashboard
SFO_dashboard = pn.Tabs( 
    ("Welcome", welcome_tab), 
    ("Yearly Market Analysis", market_analysis_tab),
    ("Neighborhood Analysis", neighborhood_analysis_tab)
)
```

## Serve the Panel Dashboard

```python
# Serve the# dashboard
SFO_dashboard.servable()
```
